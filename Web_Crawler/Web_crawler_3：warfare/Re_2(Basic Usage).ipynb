# re.search(pattern, string, flags=0)

#pattern: re string
#string: 待匹配string
#flags: 控制标记
    1. re.I (IGNORECASE) 忽略大小写 进行匹配
    2. re.M (MULTILINE)  re中的“^”能将字符串每行开始当做匹配开始
    3. re.S (DOTALL)     re中“.”匹配所有字符，默认情况下是不能匹配换行符的

```python
import re
match = re.search(r'[1-9]\d{5}', 'BIT 210008')
match1 = re.search(r'^[1-9]\d{5}', 'BIT 210008')
match2 = re.search(r'[1-9]\d{5}$', 'BIT 210008')
```

```python
if match: print(match.group(0))
else: print('No match')
```

```python
if match1: print(match1.group(0))
else: print('No match')
```

```python
if match2: print(match2.group(0))
else: print('No match')
```

# re.match(pattern, string, flag=0)

```python
import re
match = re.match(r'[1-9]\d{5}', 'ansdvibsnbd210008')
```

```python
if match: print(match.group(0))
else: print('No match')
```

# re.findall(pattern, string, flag=0)

```python
import re
ls = re.findall(r'\d.\d', '514561.15.1561454.516')
```

```python
if ls: 
    for i in ls: print(i)
else: print('No match')
```

# re.split(pattern, string, maxsplit=0, flags=0)

#maxsplit: 最大分割数

```python
import re 
re.split(r'[1-9]\d{5}', 'hello111111world222222nice333333hahaha')
```

```python
re.split(r'[1-9]\d{5}', 'hello111111world222222nice333333hahaha', maxsplit=2)
```

# re.finditer(pattern, string, flag=0)

```python
import re
imatch = re.finditer(r'[1-9]\d{5}', 'afwef215326ascawef211534cw')
```

```python
for match in imatch:
    if match:
        print(match.group(0))
```

# re.sub(pattern, repl, string, count=0, flags=0)

#repl：用于替换的字符串
#count：替换最大次数

```python
import re
re.sub(r'[1-9]\d{5}', '', 'afwef215326ascawef211534cw')
```
