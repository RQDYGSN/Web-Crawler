# 标签树的下行遍历

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://python123.io/ws/demo.html'
```

```python
r = requests.get(url)
```

```python
demo = r.text
```

```python
soup = BeautifulSoup(demo, "html.parser")
```

```python
soup.head
```

```python
soup.head.contents
```

```python
type(soup.head.contents)
```

```python
print(len(soup.head.contents))
print(len(soup.body.contents))
```

```python
soup.body.contents[1]
```

```python
for i in soup.body.contents:
    print(i)
```

```python
for i in soup.body.children:
    print(i)
```

```python
for i in soup.body.descendants:
    print(i)
```

# 标签的平行遍历

#注意：标签的平行遍历必须发生在同一个父亲节点下，否则不构成平行遍历关系。

#tag和string均可能构成标签树的节点。

```python
#遍历a后续节点
for i in soup.a.next_siblings:
    print(i)
```

```python
#遍历a前序节点
for i in soup.a.previous_siblings:
    print(i)
```
