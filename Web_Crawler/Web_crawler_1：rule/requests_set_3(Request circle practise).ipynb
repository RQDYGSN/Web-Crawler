```python
import requests
import time
import matplotlib.pyplot as plt
 
def getHTTPText(url):
    try:
        r = requests.get(url)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        return "产生异常"
```

```python
K=100
x=range(100)
y=[]
if __name__ == '__main__':
    url = "https://www.bilibili.com"
    start = time.time()
    for i in range( K ):
        getHTTPText(url)
        y.append(time.time()-start)
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(x, y, 'r-', label='Time(s)')
    ax.set_xlabel('Run Times')
    ax.set_ylabel('Time(s)')
    print('Well done.')
```

```python
url = "https://www.bilibili.com"
getHTTPText(url)
```

```python
print('In 100 times crawlings, each cost', round(y[-1]/K, 4), "second")
```
