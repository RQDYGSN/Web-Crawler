# request()

```python
import requests
```

```python
kv = {'key1': 'value1', 'key2': 'value2'}
url = 'http://python123.io'
```

#params url

```python
r = requests.request('GET', url, params = kv)
```

```python
print(r.url)
```

#data 

```python
r = requests.request('POST', url, data = kv)
```

```python
body = 'main content'
r = requests.request('POST', url, data = body) #
```

#json JSON

```python
r = requests.request('POST', url, json = kv) 
```

#headers url

```python
head = {'user-agent': 'Chrome/10'} #Chrome browser v.10
r = requests.request('POST', url, headers=head)
```

#cookies HTTPcookiecookiejar
#auth HTTP
#file 

```python
fs = {'file': open('data.xls', 'rb')}
r = requests.request('POST', url, file=fs)
```

#timeout s

```python
r = requests.request('GET', url, timeout=10)
```
