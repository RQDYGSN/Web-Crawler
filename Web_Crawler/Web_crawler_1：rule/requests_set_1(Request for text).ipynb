# request()

```python
import requests
```

```python
kv = {'key1': 'value1', 'key2': 'value2'}
url = 'http://python123.io'
```

#params 修改url名称参数

```python
r = requests.request('GET', url, params = kv)
```

```python
print(r.url)
```

#data 写入数据

```python
r = requests.request('POST', url, data = kv)
```

```python
body = 'main content'
r = requests.request('POST', url, data = body) #写入数据
```

#json 写入JSON格式数据

```python
r = requests.request('POST', url, json = kv) 
```

#headers 修改url头部信息

```python
head = {'user-agent': 'Chrome/10'} #Chrome browser v.10
r = requests.request('POST', url, headers=head)
```

#cookies 从HTTP中解析cookie，可以使字典或cookiejar类型
#auth 元祖，支持HTTP认证功能
#file 字典类型，向服务器传输文件

```python
fs = {'file': open('data.xls', 'rb')}
r = requests.request('POST', url, file=fs)
```

#timeout 设定超时时间，以s为单位

```python
r = requests.request('GET', url, timeout=10)
```

#proxies 字典类型，设定访问代理服务器，可以增加登录认证，有效隐藏源服务器地址,防止对爬虫的逆追踪。

```python
# pxs = {'http': 'http://user:pass@121.233.226.163:9999',
#        'https': 'https://121.233.226.163'}
pxs = {'http': 'https://121.233.226.163:9999',
       'https': 'https://121.233.226.163:9999'}
```

```python
r = requests.request('GET', 'https://www.baidu.com', proxies=pxs)
```

#allow_redirects True(默认) or False, 重定向开关

#stream True(默认) or False，获取内容是否立即下载

#verify True(默认) or False，是否认证SSL证书

#cert 本地SSL证书路径（SSL：Secure Sockets Layer 安全套接层）
