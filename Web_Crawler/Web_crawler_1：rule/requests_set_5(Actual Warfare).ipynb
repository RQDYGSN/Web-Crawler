
#360搜索代码框架体如下：

```python
import requests
url = 'http://www.so.com/s'

def getHTMLText(url):
    try:
        kv = {'q':'Python'}
        r = requests.get(url, params=kv)
        r.raise_for_status
        r.encoding = r.apparent_encoding
        print(r.text[2000:3000])
    except:
        print('Error!')
```

```python
getHTMLText(url)
```

#完整代码框架：

```python
import requests
import os
url = 'https://graph.baidu.com/resource/212818a69f57ce1045be501564041561.jpg?idctag=nj'
root = 'H:\\Webcrawler_file\\'
#fig_addr = root + url.split('/')[-1]
fig_addr = 'H:\\Webcrawler_file\\212818a69f57ce1045be501564041561.jpg'
try:
    if not os.path.exists(root):
        os.mkdir(root)
    if not os.path.exists(fig_addr):
        r  = requests.get(url)
        with open(fig_addr, 'wb') as f:
            f.write(r.content)
            f.close() #???不自动关闭？？？
            print('File saved well.')
    else:
        print('File has been existed.')
except:
    print('Crawler has failed.')
```

IP地址所属地查询的代码框架：

```python
import requests
url = 'http://m.ip138.com/ip.asp?ip='
ip = '58.213.113.87'
try:
    r = requests.get(url+ip)
    r.raise_for_status
    r.encoding = r.apparent_encoding
    print(r.text[1917:2022])
except:
    print('Error!')
```
